Escriba una expresión regular para corregir este problema. En particular, su expresión regular debe ser capaz de sustituir el valor `$1.456,76` por el valor `1456.76` y el valor `$56,15` por el valor `56.15`. Utilice estos (y otros) ejemplos para corroborar su funcionamiento. Apóyese en <a href="https://regex101.com/">regex101.com</a>

Vea los nombres de columna de sus tablas. Escriba una función que reciba una palabra en CamelCase y devuelva la misma palabra en snake_case. Utilice este formato como referencia:

```python
def camel_to_snake(df):
    lst = list(df.columns)
    
    new_lst = []
    
    for col in lst:
        new_lst.append(re.sub(r"(?<=[a-z])([A-Z])", r"_\1", col).lower())
    
    df.columns = new_lst
    return(df)
```

```python
camel_to_snake(sales)
```
